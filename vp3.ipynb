```python
# toto ale nebude fungovat napr. pro hodnotu 118 - jak jsme si ukazali na hodine, "zkusi se" oba ify - splni se podminka prvniho, nasledne se zkusi
# druhy if, ktery se nesplni, hodnota tedy spadne do chlivku "else". toto vyresime dalsim keyword "elif", ktery se zkusi pouze v pripade, ze prvni
# if/predchazejici elif se nesplni. else se vzdy vztahuje k bloku podminek k poslednimu "if"

Hb = float(input("Jaká je hodnota hemoglobinu v g/l?"))
if Hb < 120:
    print("Anemie :/")
elif Hb > 160:
    print("To je zas moc:/")
else:
    print("V normě :)")
```

```python
# ify, elify a else jsme si nasledne ukazali na ukolu s hodnotou BMI:

bmi = float(input("BMI: "))

if bmi < 18.5:
    print("podváha")
elif 18.5 <= bmi <= 25:
    print("norma")
elif 25 < bmi <= 30:
    print("nadváha")
else:
    print("obezita")
```

```python
""" bmi.py """
# na zaver jsme si zkouseli zadefinovat vlastni funkci pomoci keyword "def", vysvetlili pojem scope (promenna zadefinovana ve scope definice
# funkce existuje pouze v tomto prostoru), pouzili keyword "return" ke vraceni hodnoty dane funkce

def bmi(vyska, vaha):
    bmi = vaha / (vyska * vyska)
    return bmi
```

```python
# spravna definice funkce se vsim vsudy: specifikujeme datove typy vstuonich parametru i navratove hodnoty, doplnujeme dokumentaci
# funkce popisem a jeste jsme si dovolili vysledek zaokrouhlit na dve desetinna mista:

def vypocet_bmi(vyska: float, vaha: float) -> float:
    """ funkce vracejici hodnotu BMI zaokrouhlenou na 2 desetinna mista """
    bmi = vaha / (vyska*vyska)
    bmi = round(bmi, ndigits=2)
    return bmi
```
