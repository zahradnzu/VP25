### Zdrojovy kod k 3. hodine VP
Na teto hodine jsme si ukazali formatovane stringy, podminky (neboli "ify") a zkouseli definovat vlastni funkci.

```python
""" name.py """
# pomoci pismene 'f' pred string rikame, ze se bude jednat o formatovany string, coz v praxi znamena, ze pokud do nej vepiseme jmeno
# promenne v chlupatych zavorkach {}, dosadi do stringu hodnotu dane promenne.

name = "Anicka"
age = 23
veta = f"Jmenuji se {name} a je mi {age} let."
print(veta)
```

```python
# pokud bychom pred dany string pismeno f nevepsali, bude nas python brat doslova a chlupate zavorky vnimat jako jakykoliv jiny, bezny znak,
# a do vysledneho stringu hodnotu danych promennych nedosadi.

name = "Anicka"
age = 23
veta = "Jmenuji se {name} a je mi {age} let."
print(veta)
```
