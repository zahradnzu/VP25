### Zdrojovy kod k 4. hodine VP
Tuto hodinu jsme si doplnili neco malo k logickym (and, or) a komparativnim operatorum (==, >, <, >=, <=). Dominantnim tematem byly cykly (while, for) a na zaver jsme se alespon pokusili nainstalovat knihovny k analyze dat v Pythonu a ukazali si koncept Jupyter notebooku.

```python
""" anemie.py """
# v minulem skriptu resici anemii jsme nezohlednovali pohlavi, ukazali jsme si tedy jeden ze zpusobu, jak tento problem resit pomoci
# logickeho operatoru and. podminka se splni pouze tehdy, pokud jsou oba vyrazy True (plati jak podminka s Hb, tak s pohlavim).
# povsimnete si take, ze pro srovnani hodnot stringu (sex == "f") se vyuziva dvojite =, protoze jedno = je znakem prirazeni do promenne!

Hb = float(input("Hodnota hemoglobinu v g/l? "))
sex = input("Pohlavi ve formatu M/F: "). lower()

if Hb < 120 and sex == "f":
    print("anemie")
elif Hb > 160 and sex == "f":
    print("to je zas moc")
elif Hb < 135 and sex == "m":
    print("anemie")
elif Hb > 170 and sex == "m":
    print("to je zas moc")
else:
    print("v normě :)")
```

```python
# tohle reseni je kazdopadne pomerne neelegantni, protoze pro stejny vystup ("anemie" nebo "to je zas moc") pouzivame vicero podminek
# proto zkusime upravu s logickym operatorem "or", kdy podminka se splni, pokud alespon jeden z vyrazu je pravdivy:

Hb = float(input("VLožte hodnotu hemoglobinu v g/l: "))
sex = input("Jaké je pohlaví pacienta (M/F)? ").lower()

if (Hb < 120 and sex == "f") or (Hb < 135 and sex == "m"):
    print("Anemie :/")
elif (Hb > 160 and sex == "f") or (Hb > 175 and sex == "m"):
    print("To je zas moc!")
else:
    print("V normě :)")

# pozn.: operator and prakticky "lepi" dva vyrazy k sobe, zavorky tedy nejsou nutne, je lepsi je vsak po prehlednost pouzivat, protoze
# se clovek pomerne snadno ztrati, pokud konstruuje slozitejsi vyroky
```

```python
""" mnau.py """
# na prikladu skriptu jehoz input je pocet mnouknuti a output zamnkoukani tolikrat, kolikrat jsme urcili jsme si ukazali princip while loopu
# u konstruovani while loopu je nutne mit na pameti, ze by mel nekdy skoncit a mit tak osetreny vsechny varianty

# !! nasledujici cyklus je nekonecny a poustet ho je nebezpecne
pocet_mnouknuti = int(input("Kolikrat mam mnouknout? "))
pocitadlo = 0

while pocet_mnouknuti > pocitadlo:
    print("mňau")
```
