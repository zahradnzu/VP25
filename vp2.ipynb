```python
""" meds.py """
# na prikladu seznamu medikace jsme si ukazali princip indexovani a 'sahani' po jednotlivych 'podjednotkach' dane promenne
# kazda z nich ma svuj tzv. index, zjednodusene lze chapat jako 'poradove cislo', diky kteremu s ni muzeme dale nakladat
# konkretne v Pythonu se indexuje od 0, tj. prvni polozka v seznamu (Eliquis) ma index 0, druha 1, treti 2 a tak dale

# nasledujici kod zadefinuje promennou seznam_leku typu list, obsahuje 5 'polozek' /stringu, tj. textovych retezcu/, pomoci
# funkce print() si zobrazime prvni polozku pod indexem O. zapis je pomoci hranatych zavorek []
seznam_leku = ["Eliquis", "Ritalin", "Prenessa", "Atoris", "Elicea"]
print(seznam_leku[0])
```

```python
# pomoci indexu minus n muzeme take sahnout na polozky odzadu, napr. pod indexem -1 je polozka posledni:
print(seznam_leku[-1])
```

```python
# lze take rozbrazit rozsah pomoci :, nutno pamatovat, ze prvni cislo je inkluzivni, druhe exkluzivni (polozka pod timto indexem jiz nebude zahrnuta)
print(seznam_leku[1:3])
```

```python
# pokud chceme vybrat od nejakeho indexu az do konce, lze take alternativne nasledovne (nezadavame konecny index):
print(seznam_leku[2:])
```

```python
""" mail.py """
# FUNKCE/METODY. na prikladu praci s *hypotetickou* mailovou adresou jsme si vyzkouseli pouziti tzv. metod, zjednodusenne je lze chapat jako
# funkce zavesene na konkretni objekt, ktery musi byt jasne daneho datoveho typu /napr. metodu na zvetseni prvniho pismenka stringu nelze
# zavesit na integer.
# Take jsme si ukazali funkci input, jejiz poslanim je sebrat vstup od uzivatele z terminalu a ev. ho ulozit jako string. 

mail = input("Zde vlozte svou mailovou adresu: ")
print(mail)
```

```python
# ukazali jsme si metodu .strip(), ktera zbavi textovy retezec mezer pred prvnim znakem a mezer po poslednim znaku
# alternativne lze pouzit .lstrip() ci .rstrip(), pokud chceme string zbavit mezer jen vlevo ci vpravo

mail = input("Zde vlozte svou mailovou adresu: ")
ocisteny_mail = mail.strip()
print(ocisteny_mail) 
```

```python
# zde nutne neni potreba vytvaret novou promennou - metodu lze 'navesit' na input(), aplikuje se tedy na navratovou hodnotu
# funkce input()

mail = input("Zde vlozte svou mailovou adresu: ").strip()
print(mail)
```

```python
# co delat v pripade, ze by mezery byly i veprostred vlozene mailove adresy? pro tento problem existuje metoda .replace(),
# ktera bere dva argumenty: prvni je string "old", tedy sekvenci znaku kterou hledame a chceme nahradit, a druhy je string
# "new", kterym specifikujeme cim ho chceme nahradit. v nasem pripade nechceme nahrazovat nicim, vkladame tedy prazdny string

mail = mail.replace(" ", "")
print(mail)
```

```python
# mimo jine jsme narazili na kosmeticky problem, ze vlastne nevime, *co* tzv. printime do terminalu. printu muzeme pridat
# dalsi argument, vysledek vyprinti oddeleny mezerou:

print("Mailova adresa je:", mail)
```

```python
# pokud by se nam nelibilo oddeleni jednotlivych argumentu mezerou, muzeme specifikovat nepovinny parametr "sep", tedy separator:

print("Mailova adresa je:", mail, sep="AAAAAAAAAAAAAAAAAAA")
```
