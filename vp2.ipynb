```python
""" meds.py """
# na prikladu seznamu medikace jsme si ukazali princip indexovani a 'sahani' po jednotlivych 'podjednotkach' dane promenne
# kazda z nich ma svuj tzv. index, zjednodusene lze chapat jako 'poradove cislo', diky kteremu s ni muzeme dale nakladat
# konkretne v Pythonu se indexuje od 0, tj. prvni polozka v seznamu (Eliquis) ma index 0, druha 1, treti 2 a tak dale

# nasledujici kod zadefinuje promennou seznam_leku typu list, obsahuje 5 'polozek' /stringu, tj. textovych retezcu/, pomoci
# funkce print() si zobrazime prvni polozku pod indexem O. zapis je pomoci hranatych zavorek []
seznam_leku = ["Eliquis", "Ritalin", "Prenessa", "Atoris", "Elicea"]
print(seznam_leku[0])
```

```python
# pomoci indexu minus n muzeme take sahnout na polozky odzadu, napr. pod indexem -1 je polozka posledni:
print(seznam_leku[-1])
```

```python
# lze take rozbrazit rozsah pomoci :, nutno pamatovat, ze prvni cislo je inkluzivni, druhe exkluzivni (polozka pod timto indexem jiz nebude zahrnuta)
print(seznam_leku[1:3])
```

```python
# pokud chceme vybrat od nejakeho indexu az do konce, lze take alternativne nasledovne (nezadavame konecny index):
print(seznam_leku[2:])
```
