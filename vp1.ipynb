```python
""" kalkulacka.py """
# PROMENNE, FUNKCE PRINT
# zde jsme si ukazali koncept promenne, do ktere prirazujeme pomoci znaminka = pozadovanou hodnotu, zde do promenne x prirazujeme hodnotu
# 2, do promenne y hodnotu 3 a pomoci print() ukazujeme jejich soucet

x = 2
y = 3
print(x + y)
```

```python
# alternativne:
x = 2
y = 3
vysledek = x + y
print(vysledek)
```

```python
# demonstrace rozdilu string (textovy retezec) a int (cele cislo) - v Pythonu lze "scitat" stringy, dojde k jejich slepeni dohromady,
# nejedna se o soucet a ve vetsine programovacich jazyku je tato operace nemozna
x = "2"
y = "3"
vysledek = x + y
print(vysledek)
```
